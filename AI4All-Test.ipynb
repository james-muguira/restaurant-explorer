```python
import pandas as pd
```

```python
Test_data=pd.read_csv("test_data.csv")
```

```python
import boto3,sys

sm_rt = boto3.Session().client('runtime.sagemaker')
```

```python
ep_name="AIForAll-Aqua-RevPredi"
```

```python
prediction_arr=[]
with open("test_data.csv") as f:
    lines = f.readlines()
    for l in lines[1:]:
        print(l)
        response = sm_rt.invoke_endpoint(EndpointName = ep_name, ContentType = 'text/csv', Accept = 'text/csv', Body=l)
        response = response['Body'].read().decode("utf-8")
        print("Predicted Revenue - ", response, "\n")
        prediction_arr.append(response)
```

```python
Test_data['predicted_revenue'] = prediction_arr
Test_data
```

```python
Test_data.to_csv("Test_data_predictions.csv")
```
