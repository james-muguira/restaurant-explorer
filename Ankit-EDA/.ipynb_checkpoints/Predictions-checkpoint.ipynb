```python
# Predict of test set

import boto3, sys

sm_rt = boto3.Session().client('runtime.sagemaker')
```

```python
endpoint_name = "TEAMPURPLE-2021-09-01"
```

```python
prediction_arr = []
```

```python
with open("test_data.csv") as f:
    lines = f.readlines()
    for l in lines[1:]:
        print(l)
        response = sm_rt.invoke_endpoint(EndpointName = endpoint_name, ContentType = 'text/csv', Accept = 'text/csv', Body=l)
        response = response['Body'].read().decode("utf-8")
        print("Predicted Revenue - ", response, "\n")
        prediction_arr.append(response)
```

```python
test_data['predicted_revenue'] = prediction_arr
```

```python
test_data
```

```python
test_data.to_csv("predictions.csv")
```
